```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import urllib.request as urequest
```

```python
URL = 'https://en.wikipedia.org/wiki/Economy_of_the_United_States'

request = urequest.Request(URL)

response = urequest.urlopen(request)
```

```python
data_GDP = response.read()
data_GDP
```

```python
soup = BeautifulSoup(data_GDP, 'html.parser')
soup
```

```python
data = soup.find_all('table', {'class':'wikitable'})
print (f'Number of wikitable tables found: {len(data)}')
```

```python
if data:
    table = data[0]
    df = pd.read_html(str(table))[0]
```

### Some analysis may require that we work on the numbers that form the percentages shared within a column, how do we get the numbers without the `%` sign.

```python
df['GDP growth (real)']
```

```python
df['GDP growth (real)'] = df['GDP growth (real)'].str.replace('%', '').astype(float)
df['GDP growth (real)']
```

```python
df['Inflation rate (in Percent)'] = df['Inflation rate (in Percent)'] .str.replace('%', '').astype(float)
df['Inflation rate (in Percent)'] 
```
